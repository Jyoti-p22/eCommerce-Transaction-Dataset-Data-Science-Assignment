```python
# import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import numpy as np
```

# Customer Data Analysis

```python
customers_df = pd.read_csv('C:/Users/Admin/Desktop/Customers.csv')
```

```python
# check top 5 rows 
customers_df.head()
```

```python
#describe the data
customers_df.describe()
```

```python
# check info about customer
customers_df.info()
```

```python
# check columns
customers_df.columns
```

```python
customers_df.shape
```

```python
# check for missing values
customers_df.isnull().sum() 
```

```python
# Regional Distribution
# Customer Distribution by Region
region_dist = customers_df['Region'].value_counts()
region_dist
```

```python
# Signup Date Analysis
customers_df['SignupDate'] = pd.to_datetime(customers_df['SignupDate'])
customers_df['SignupYear'] = customers_df['SignupDate'].dt.year
customers_df['SignupMonth'] = customers_df['SignupDate'].dt.month

print("Signup Date Analysis:")
print("Earliest Signup:", customers_df['SignupDate'].min())
print("Latest Signup:", customers_df['SignupDate'].max())
print("\nSignups by Year:")
print(customers_df['SignupYear'].value_counts().sort_index())
  
```

```python
# Visualizations for Customer Analysis
plt.figure(figsize=(15, 10))

# Regional Distribution
plt.subplot(2, 2, 1)
sns.barplot(x=region_dist.index, y=region_dist.values)
plt.title('Customer Distribution by Region')
plt.xticks(rotation=45)
```

```python
# Visualizations for Customer Analysis
plt.figure(figsize=(15, 10))

# Signup Trend Over Time
plt.subplot(2, 2, 2)
customers_df['SignupDate'].dt.date.value_counts().sort_index().plot()
plt.title('Customer Signup Trend Over Time')
plt.xticks(rotation=45)
```

```python
# Visualizations for Customer Analysis
plt.figure(figsize=(15, 10))

# Monthly Signup Pattern
plt.subplot(2, 2, 3)
sns.countplot(data=customers_df, x='SignupMonth')
plt.title('Monthly Signup Pattern')
plt.xlabel('Month')
plt.ylabel('Number of Signups')
```

```python
# Visualizations for Customer Analysis
plt.figure(figsize=(15, 10))

#Yearly Signup Pattern
plt.subplot(2, 2, 4)
sns.countplot(data=customers_df, x='SignupYear')
plt.title('Yearly Signup Pattern')
plt.xlabel('Year')
plt.ylabel('Number of Signups')
```

```python
plt.tight_layout()
plt.savefig('customer_analysis.png')
plt.close()
```

```python
# Category Analysis
print("\n3. Product Categories Distribution:")
category_dist = products_df['Category'].value_counts()
print(category_dist)
```

```python
# 4. Price Analysis
print("\n4. Price Analysis:")
print("Price Statistics by Category:")
print(products_df.groupby('Category')['Price'].describe())
```

```python
# Visualizations for Product Analysis
plt.figure(figsize=(15, 10))

# Category Distribution
plt.subplot(2, 2, 1)
sns.barplot(x=category_dist.index, y=category_dist.values)
plt.title('Products by Category')
plt.xticks(rotation=45)
```

```python
# Visualizations for Product Analysis
plt.figure(figsize=(15, 10))

# Price Distribution
plt.subplot(2, 2, 2)
sns.histplot(products_df['Price'], bins=30)
plt.title('Price Distribution')
```

```python
# Visualizations for Product Analysis
plt.figure(figsize=(15, 10))

# Price Box Plot by Category
plt.subplot(2, 2, 3)
sns.boxplot(x='Category', y='Price', data=products_df)
plt.title('Price Range by Category')
plt.xticks(rotation=45)
```

```python
# Visualizations for Product Analysis
plt.figure(figsize=(15, 10))

# Plot 4: Average Price by Category
plt.subplot(2, 2, 4)
avg_price = products_df.groupby('Category')['Price'].mean().sort_values(ascending=False)
sns.barplot(x=avg_price.index, y=avg_price.values)
plt.title('Average Price by Category')
plt.xticks(rotation=45)
```

```python
plt.tight_layout()
plt.savefig('product_analysis.png')
plt.close()
```

# Transaction Data Analysis

```python
# load the dataset
transactions_df = pd.read_csv('C:/Users/Admin/Desktop/transactions.csv')
```

```python
# check top 5 rows 
transactions_df.head()
```

```python
#describe the data
transactions_df.describe()
```

```python
# check info about customer
transactions_df.info()
```

```python
transactions_df.columns
```

```python
transactions_df.shape
```

```python
transactions_df.isnull().sum() 
```

```python
# Transaction Date Analysis
transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'])
transactions_df['TransactionYear'] = transactions_df['TransactionDate'].dt.year
transactions_df['TransactionMonth'] = transactions_df['TransactionDate'].dt.month

print("Transaction Date Analysis:")
print("Earliest Transaction:", transactions_df['TransactionDate'].min())
print("Latest Transaction:", transactions_df['TransactionDate'].max())
print("\nTransactions by Year:")
print(transactions_df['TransactionYear'].value_counts().sort_index())
```

```python
# Quantity Analysis
print("Quantity Analysis:")
print("Quantity Statistics:")
print(transactions_df['Quantity'].describe())
```

```python
# Customer Purchase Analysis
print("Customer Purchase Analysis:")
customer_purchases = transactions_df['CustomerID'].value_counts()
print("Average purchases per customer:", customer_purchases.mean())
print("Maximum purchases by a customer:", customer_purchases.max())
```

```python
# Visualizations for Transaction Analysis
plt.figure(figsize=(15, 10))

# Transaction Trend Over Time
plt.subplot(2, 2, 1)
transactions_df['TransactionDate'].dt.date.value_counts().sort_index().plot()
plt.title('Transaction Trend Over Time')
plt.xticks(rotation=45)
```

```python
# Visualizations for Transaction Analysis
plt.figure(figsize=(15, 10))

# Monthly Transaction Pattern
plt.subplot(2, 2, 2)
sns.countplot(data=transactions_df, x='TransactionMonth')
plt.title('Monthly Transaction Pattern')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
```

```python
# Visualizations for Transaction Analysis
plt.figure(figsize=(15, 10))

# Quantity Distribution
plt.subplot(2, 2, 3)
sns.histplot(transactions_df['Quantity'], bins=30)
plt.title('Quantity Distribution')
```

```python
# Visualizations for Transaction Analysis
plt.figure(figsize=(15, 10))

# Customer Purchase Frequency
plt.subplot(2, 2, 4)
sns.histplot(customer_purchases.values, bins=30)
plt.title('Customer Purchase Frequency')
plt.xlabel('Number of Purchases')
plt.ylabel('Number of Customers')
```

```python
plt.tight_layout()
plt.savefig('transaction_analysis.png')
plt.close()
```

```python
# Additional Transaction Insights
print("Additional Transaction Insights:")
print("Average items per transaction:", transactions_df['Quantity'].mean())
print("Most common quantity ordered:", transactions_df['Quantity'].mode()[0])
print("Number of unique customers:", transactions_df['CustomerID'].nunique())
print("Number of unique products:", transactions_df['ProductID'].nunique())
```
